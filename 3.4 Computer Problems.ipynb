```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline, interp1d, splrep, splev
```

```python
#3.4.1.a)
x = np.array([0, 1, 2, 3,])
y = np.array([3, 5, 4, 1])
cs = CubicSpline(x,y, bc_type='natural')
xs = np.arange(-0.5, 4, 0.1)
plt.figure(figsize=(6.5,4))
plt.plot(x, y, 'o', label='data')
plt.plot(xs, cs(xs), label='S')
plt.legend(loc='lower left', ncol=1)
plt.show()
```

```python
#3.4.11.a)
x = np.array([1960, 1970, 1990, 2000])
y = np.array([3039585530, 3707475887, 5281653820, 6079603571])
cs = CubicSpline(x, y, bc_type='natural')
xs = np.arange(1960, 2001, 1)
plt.figure(figsize=(6.5, 4))
plt.plot(x, y, 'o', label='data')
plt.plot(xs, cs(xs), label='spline')
plt.legend(loc='lower left', ncol=2)
plt.show()

print('1980 World Population:', cs(1980))
```

```python
#3.4.11.b)
x = np.array([1960, 1970, 1990, 2000])
y = np.array([3039585530, 3707475887, 5281653820, 6079603571])
tck = splrep(x, y, k=1, s=0)
xs = np.arange(1960, 2001, 1)
yder1 = splev(1960, tck, der=1)
yder2 = splev(2000, tck, der=1)

cs = CubicSpline(x, y, bc_type=((1, yder1), (1, yder2)))
plt.figure()
plt.plot(x, y, 'o', label='data')
plt.plot(xs, cs(xs), label='clamped spline')
plt.legend(loc='best')
plt.show()

print('1980 World Population:', cs(1980))
```

```python
#3.4.12.a)
x = np.array([1800, 1850, 1900, 2000])
y = np.array([280, 283, 291, 370])
csn = CubicSpline(x, y, bc_type='natural')
csk = CubicSpline(x, y, bc_type='not-a-knot')
xs = np.arange(1800, 2001, 1)
plt.figure(figsize=(6.5,4))
plt.plot(x, y, 'o', label='Data')
plt.plot(xs, csn(xs), label='Natural Spline')
plt.plot(xs, csk(xs), label='Not-A-Knot Spline')
plt.legend(loc='best')
plt.show()

print("Natural Estimate:", csn(1950))
print("Not-A-Knot Estimate:", csk(1950))
```

```python
#3.4.13.
x = np.arange(1994, 2004, 1)
y = np.array([67.052, 68.008, 69.803, 72.024, 73.400, 72.063, 74.669, 74.487, 74.065, 76.777])
csn = CubicSpline(x, y, bc_type='natural')
csk = CubicSpline(x, y, bc_type='not-a-knot')
#csp = CubicSpline(x, y, bc_type='parabolic')
xs = np.linspace(1994, 2003, num=50, endpoint=True)

plt.figure(figsize=(6.5,4))
plt.plot(x, y, 'o', label='Data')
plt.plot(xs, csn(xs), label='Natural Spline')
plt.plot(xs, csk(xs), label='Not-A-Knot Spline')
#plt.plot(xs, csp(xs), label='Parabolically Terminated Spline')
plt.legend(loc='best')
plt.show()
```
